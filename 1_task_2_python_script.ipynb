# Data Exploration and Cleaning

```python
import pandas as pd
from pandas import read_excel
```

## Data Set 1: Historical GDP
### Source: https://data.worldbank.org/indicator/NY.GDP.MKTP.CD

```python
gdp_ds = pd.read_excel('historical_gdp_data_source.xlsx')
gdp_ds.head()
```

```python
column_names = gdp_ds.columns
column_names
```

```python
gdp_ds.tail()
```

```python
gdp_ds = gdp_ds.drop(gdp_ds.index[-1])
```

```python
gdp_ds.tail()
```

```python
na_counts = gdp_ds.isna().sum(axis=1)
count_gt_zero = na_counts[na_counts > 0].count()
count_gt_zero
```

#### Columns with NA are removed from the dataset

```python
gdp_ds.dropna(inplace=True)
```

```python
na_counts = gdp_ds.isna().sum(axis=1)
count_gt_zero = na_counts[na_counts > 0].count()
count_gt_zero
```

```python
gdp_ds
```

```python
column_types = gdp_ds.dtypes
column_types
```

#### The data type in the years columns is transformed to integer.

```python
gdp_ds.iloc[:, list(range(1, 23))] = gdp_ds.iloc[:, list(range(1, 23))].astype(int)
```

```python
column_types = gdp_ds.dtypes
column_types
```

```python
gdp_ds
```

```python
gdp_ds.rename(columns={gdp_ds.columns[0]: 'country'}, inplace=True)
gdp_ds.head(5)
```

#### Check for duplicate values (countries).

```python
gdp_ds['country'].is_unique
```

#### The data set is cleaned, ready to be exported to the database 

```python
gdp_ds.to_csv('cleaned_datasets/historical_gdp_cleaned.csv', index=False)
```
